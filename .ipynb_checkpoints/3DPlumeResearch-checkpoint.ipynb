```python
import numpy as np
import pandas as pd
import xarray as xr
import plotly.graph_objects as go
from plotly.subplots import make_subplots
import matplotlib.pyplot as plt
from datetime import datetime, timedelta
import warnings
warnings.filterwarnings('ignore')


%matplotlib widget  
```

```python
print("ðŸ“‚ Loading HYCOM sample...")
hycom = xr.open_dataset('data/hycom_combined_march/hycom_march_2011_combined.nc')
print(f"âœ… Loaded successfully!")

print("\n=== DATASET STRUCTURE ===")
print(f"Dimensions: {dict(hycom.dims)}")
print(f"Data variables: {list(hycom.data_vars)}")
print(f"Coordinates: {list(hycom.coords)}")

# Check what we actually have
print("\n=== DIMENSION DETAILS ===")
for dim in hycom.dims:
    print(f"{dim}: {hycom[dim].values.shape} - values: {hycom[dim].values[:3] if len(hycom[dim].values) > 3 else hycom[dim].values}")

print("\n=== COORDINATE DETAILS ===")
for coord in hycom.coords:
    if coord not in hycom.dims:  # Skip dimensions we already printed
        print(f"{coord}: shape={hycom[coord].shape}, range={hycom[coord].values.min():.2f} to {hycom[coord].values.max():.2f}")
```

```python
print("\n=== RENAMING TO STANDARD NAMES ===")
# Create a copy with standard names for easier use
hycom_std = hycom.rename({
    'MT': 'time',
    'Depth': 'depth',
    'Latitude': 'lat',
    'Longitude': 'lon'
})

print(f"âœ… Renamed dimensions:")
print(f"  time: {len(hycom_std.time)} steps")
print(f"  depth: {hycom_std.depth.values} m")
print(f"  lat: {hycom_std.lat.values.min():.2f}Â°N to {hycom_std.lat.values.max():.2f}Â°N")
print(f"  lon: {hycom_std.lon.values.min():.2f}Â°E to {hycom_std.lon.values.max():.2f}Â°E")

# Convert time if needed (HYCOM often uses days since 1900-12-31)
if 'units' in hycom_std.time.attrs:
    print(f"\nTime units: {hycom_std.time.attrs['units']}")
    # Convert to datetime
    hycom_std['time'] = pd.to_datetime(hycom_std.time.values)
    print(f"First time: {hycom_std.time.values[0]}")
    print(f"Last time: {hycom_std.time.values[-1]}")
```

```python
print("\n=== FINDING FUKUSHIMA LOCATION ===")
fukushima_lat = 37.42
fukushima_lon = 141.21

# For 2D arrays, we need to find the closest point
lat_2d = hycom_std.lat.values
lon_2d = hycom_std.lon.values

# Calculate distance to all grid points
distances = np.sqrt((lat_2d - fukushima_lat)**2 + 
                    ((lon_2d - fukushima_lon) * np.cos(np.radians(fukushima_lat)))**2)

# Find minimum distance
min_idx = np.unravel_index(np.argmin(distances), distances.shape)
lat_idx, lon_idx = min_idx

print(f"Fukushima location: {fukushima_lat}Â°N, {fukushima_lon}Â°E")
print(f"Closest grid point: index ({lat_idx}, {lon_idx})")
print(f"Grid coordinates: lat={lat_2d[lat_idx, lon_idx]:.2f}Â°N, lon={lon_2d[lat_idx, lon_idx]:.2f}Â°E")
print(f"Distance: {distances[lat_idx, lon_idx]:.3f} degrees")
```

```python
print("\n=== SURFACE CURRENTS AT FUKUSHIMA ===")

# Extract data at Fukushima location
u_fukushima = hycom_std.u.isel(time=0, depth=0).values
v_fukushima = hycom_std.v.isel(time=0, depth=0).values

# Get a small region around Fukushima for visualization
lat_range = slice(max(0, lat_idx-200), min(lat_2d.shape[0], lat_idx+200))
lon_range = slice(max(0, lon_idx-200), min(lat_2d.shape[1], lon_idx+200))

fig, axes = plt.subplots(1, 2, figsize=(14, 6))

# Plot u component
im1 = axes[0].imshow(u_fukushima[lat_range, lon_range], 
                     cmap='RdBu_r', origin='lower')
axes[0].set_title('Zonal Current (u) near Fukushima [m/s]')
axes[0].set_xlabel('Longitude index')
axes[0].set_ylabel('Latitude index')
plt.colorbar(im1, ax=axes[0])

# Plot v component
im2 = axes[1].imshow(v_fukushima[lat_range, lon_range], 
                     cmap='RdBu_r', origin='lower')
axes[1].set_title('Meridional Current (v) near Fukushima [m/s]')
axes[1].set_xlabel('Longitude index')
axes[1].set_ylabel('Latitude index')
plt.colorbar(im2, ax=axes[1])

# Mark Fukushima location
fukushima_rel_x = lon_idx - lon_range.start
fukushima_rel_y = lat_idx - lat_range.start
axes[0].plot(fukushima_rel_x, fukushima_rel_y, 'r*', markersize=15, label='Fukushima')
axes[1].plot(fukushima_rel_x, fukushima_rel_y, 'r*', markersize=15, label='Fukushima')
axes[0].legend()
axes[1].legend()

plt.tight_layout()
plt.show()
```

```python
print("\n=== VERTICAL PROFILE AT FUKUSHIMA ===")

# Extract vertical profile at Fukushima
u_profile = hycom_std.u.isel(time=0, Y=lat_idx, X=lon_idx).values
v_profile = hycom_std.v.isel(time=0, Y=lat_idx, X=lon_idx).values
depths = hycom_std.depth.values

fig, axes = plt.subplots(1, 3, figsize=(15, 5))

axes[0].plot(u_profile, depths, 'b-', linewidth=2, marker='o')
axes[0].set_xlabel('Zonal Current (u) [m/s]')
axes[0].set_ylabel('Depth [m]')
axes[0].set_title('Vertical Profile - u')
axes[0].invert_yaxis()
axes[0].grid(True, alpha=0.3)

axes[1].plot(v_profile, depths, 'g-', linewidth=2, marker='o')
axes[1].set_xlabel('Meridional Current (v) [m/s]')
axes[1].set_title('Vertical Profile - v')
axes[1].invert_yaxis()
axes[1].grid(True, alpha=0.3)

# Calculate speed
speed_profile = np.sqrt(u_profile**2 + v_profile**2)
axes[2].plot(speed_profile, depths, 'r-', linewidth=2, marker='o')
axes[2].set_xlabel('Current Speed [m/s]')
axes[2].set_title('Vertical Profile - Speed')
axes[2].invert_yaxis()
axes[2].grid(True, alpha=0.3)

plt.tight_layout()
plt.show()

print(f"\nSurface current at Fukushima: u={u_profile[0]:.4f} m/s, v={v_profile[0]:.4f} m/s")
print(f"Bottom current (1000m): u={u_profile[-1]:.4f} m/s, v={v_profile[-1]:.4f} m/s")
```

```python
# Create a 10x10 grid in the deep Pacific
test_lats = np.linspace(25.0, 35.0, 10)  # 25Â°N to 35Â°N
test_lons = np.linspace(160.0, 180.0, 10)  # 160Â°E to 180Â°E

# Count valid depths at each location
valid_counts = np.zeros((10, 10))

print("\nTesting grid (showing number of valid depth levels):")
print("Lat\\Lon", end="")
for lon in test_lons:
    print(f" | {lon:6.1f}", end="")
print("\n" + "-" * 80)

for i, lat in enumerate(test_lats):
    print(f"{lat:6.1f}", end="")
    for j, lon in enumerate(test_lons):
        # Find closest grid point
        distances = np.sqrt((lat_2d - lat)**2 + 
                          ((lon_2d - lon) * np.cos(np.radians(lat)))**2)
        y_idx, x_idx = np.unravel_index(np.argmin(distances), lat_2d.shape)
        
        # Count valid depths
        valid_count = 0
        for d_idx in range(len(depths)):
            u_val = hycom_std.u.isel(time=0, depth=d_idx, Y=y_idx, X=x_idx).values
            if not np.isnan(u_val):
                valid_count += 1
        
        valid_counts[i, j] = valid_count
        print(f" | {valid_count:6d}", end="")
    print()
```

```python
# We want a location that:
# 1. Has full depth data (all 6 levels)
# 2. Is reasonably close to Fukushima
# 3. Is in the Kuroshio Current path

# Define search area (offshore Japan)
search_lats = np.linspace(30.0, 40.0, 20)  # 30Â°N to 40Â°N
search_lons = np.linspace(140.0, 150.0, 20)  # 140Â°E to 150Â°E

print(f"Searching {len(search_lats)}x{len(search_lons)} grid near Japan...")

# Find locations with full depth data AND strong currents
best_score = -1
best_location = None
best_indices = None

for i, lat in enumerate(search_lats):
    for j, lon in enumerate(search_lons):
        # Find closest grid point
        distances = np.sqrt((lat_2d - lat)**2 + 
                           ((lon_2d - lon) * np.cos(np.radians(lat)))**2)
        y_idx, x_idx = np.unravel_index(np.argmin(distances), lat_2d.shape)
        
        # Check if all depths have data
        all_valid = True
        for d_idx in range(len(depths)):
            u_val = hycom_std.u.isel(time=0, depth=d_idx, Y=y_idx, X=x_idx).values
            if np.isnan(u_val):
                all_valid = False
                break
        
        if all_valid:
            # Calculate surface current speed (indicator of Kuroshio)
            u_surf = hycom_std.u.isel(time=0, depth=0, Y=y_idx, X=x_idx).values
            v_surf = hycom_std.v.isel(time=0, depth=0, Y=y_idx, X=x_idx).values
            speed = np.sqrt(u_surf**2 + v_surf**2)
            
            # Calculate distance from Fukushima
            dist_from_fukushima = np.sqrt((lat - 37.42)**2 + 
                                         ((lon - 141.02) * np.cos(np.radians(37.42)))**2)
            
            # Score: high current speed, close to Fukushima
            score = speed / (dist_from_fukushima + 1.0)
            
            if score > best_score:
                best_score = score
                best_location = (lat, lon)
                best_indices = (y_idx, x_idx)

print(f"\nâœ… Optimal release location found!")
print(f"   Latitude: {best_location[0]:.3f}Â°N")
print(f"   Longitude: {best_location[1]:.3f}Â°E")
print(f"   Grid indices: ({best_indices[0]}, {best_indices[1]})")

# Get currents at this location
print("\n=== CURRENTS AT OPTIMAL LOCATION ===")
print("Depth [m] | u [m/s] | v [m/s] | Speed [m/s]")
print("-" * 45)

for d_idx, depth in enumerate(depths):
    u_val = hycom_std.u.isel(time=0, depth=d_idx, Y=best_indices[0], X=best_indices[1]).values
    v_val = hycom_std.v.isel(time=0, depth=d_idx, Y=best_indices[0], X=best_indices[1]).values
    speed = np.sqrt(u_val**2 + v_val**2)
    print(f"{depth:9.0f} | {u_val:8.4f} | {v_val:8.4f} | {speed:10.4f}")
```

```python
# ============================================
# PROFESSIONAL 3D PARTICLE VISUALIZATION SYSTEM
# UPDATED WITH CARTOPY (NO BASEMAP DEPENDENCY)
# ============================================
print("\n" + "="*80)
print("ðŸ”¬ PROFESSIONAL 3D PARTICLE VISUALIZATION SYSTEM")
print("MODERN VERSION WITH CARTOPY")
print("="*80)

import numpy as np
import matplotlib.pyplot as plt
from matplotlib import cm
from matplotlib.patches import Rectangle
import matplotlib.animation as animation
from mpl_toolkits.mplot3d import Axes3D
import cartopy.crs as ccrs
import cartopy.feature as cfeature
import ipywidgets as widgets
from IPython.display import display, clear_output, HTML
import plotly.graph_objects as go
from plotly.subplots import make_subplots
import xarray as xr
import netCDF4 as nc
import json
import time
from datetime import datetime, timedelta
import warnings
warnings.filterwarnings('ignore')

# ============================================
# 1. DIRECT PORT OF YOUR PARTICLEENGINE.JS LOGIC
# ============================================
print("\nðŸ“¦ Loading your ParticleEngine physics from web/particleEngine.js...")

class ParticleEngine3D:
    """Direct Python port of your web/particleEngine.js with CMEMS EKE integration"""
    
    def __init__(self, n_particles=10000):
        # From your constructor in particleEngine.js
        self.n_particles = n_particles
        self.particles = None
        self.currents = None
        self.eke_data = None
        self.bathymetry = None
        
        # From your init() method
        self.simulation_day = 0
        self.max_days = 365 * 3  # 3 years simulation
        self.time_step = 3600  # 1 hour in seconds
        self.substeps = 4  # From your RK4 implementation
        
        # Physics parameters (from your web controls)
        self.params = {
            'diffusion_coefficient': 10.0,  # mÂ²/s
            'vertical_diffusion': 0.001,    # mÂ²/s
            'eke_enhancement': 1.5,         # EKE scaling factor
            'vertical_mixing': 0.1,         # Vertical mixing strength
            'subsurface_decay': 0.95,       # Subsurface retention
            'surface_evaporation': 0.01,    # Surface loss rate
            'advection_strength': 1.0       # Current strength multiplier
        }
        
        # Fukushima release location
        self.release_lat = 35.789
        self.release_lon = 142.105
        self.release_depth = -50  # 50m depth
        
        print(f"âœ… ParticleEngine3D initialized for {n_particles} particles")
        print(f"ðŸ“ Release at {self.release_lat}Â°N, {self.release_lon}Â°E, {abs(self.release_depth)}m depth")
    
    def load_hycom_data(self):
        """Load HYCOM current data from NetCDF"""
        HYCOM_PATH = "/Users/shuian/PycharmProjects/Fukushima_Plume_Simulator/data/hycom_combined_march/hycom_march_2011_combined.nc"
        print(f"ðŸ“¥ Loading HYCOM currents from {HYCOM_PATH}")

        try:
            ds = xr.open_dataset(HYCOM_PATH)
            self.currents = {
                'u': ds['u'].values,      # Eastward velocity
                'v': ds['v'].values,      # Northward velocity  
                'w': ds['w'].values,      # Vertical velocity
                'lat': ds['lat'].values,
                'lon': ds['lon'].values,
                'depth': ds['depth'].values,
                'time': ds['time'].values
            }
            print(f"âœ… HYCOM data loaded: {self.currents['u'].shape}")
            return True
        except Exception as e:
            print(f"âŒ Failed to load HYCOM: {e}")
            return False
    
    def load_eke_data(self):
        """Load the interpolated 3D EKE data"""
        print("ðŸŒ€ Loading interpolated 3D EKE...")
        
        EKE_PATH = "/Users/shuian/PycharmProjects/Fukushima_Plume_Simulator/archive/cmems_EKE_data/eke_interpolated_to_3d_hycom.nc"
        
        try:
            self.eke_data = xr.open_dataset(EKE_PATH)
            print(f"âœ… Loaded 3D EKE: {EKE_PATH}")
            print(f"   Shape: {self.eke_data.eke.shape}")
            print(f"   EKE range: {self.eke_data.eke.min().values:.2e} to {self.eke_data.eke.max().values:.2e}")
            
            # Optional: Verify it matches HYCOM
            if hasattr(self, 'hycom_data'):
                print(f"   HYCOM match: {self.hycom_data.u.shape == self.eke_data.eke.shape}")
                
        except Exception as e:
            print(f"âŒ Error loading EKE: {e}")
            print("   Using synthetic EKE...")
            self._create_synthetic_eke()
    
    def load_bathymetry(self, filepath):
        """Load bathymetry data"""
        print(f"ðŸ“¥ Loading bathymetry from {filepath}")
        try:
            ds = xr.open_dataset(filepath)
            self.bathymetry = {
                'elevation': ds['elevation'].values,
                'lat': ds['lat'].values,
                'lon': ds['lon'].values
            }
            print(f"âœ… Bathymetry loaded: {self.bathymetry['elevation'].shape}")
            return True
        except Exception as e:
            print(f"âŒ Failed to load bathymetry: {e}")
            return False
    
    def initialize_particles(self):
        """Initialize particles at release location with random spread"""
        print(f"ðŸŽ¯ Initializing {self.n_particles} particles...")
        
        # Create particle array: [lat, lon, depth, age, active, concentration]
        self.particles = np.zeros((self.n_particles, 6))
        
        # Set initial positions (small random spread around release point)
        self.particles[:, 0] = self.release_lat + np.random.normal(0, 0.01, self.n_particles)
        self.particles[:, 1] = self.release_lon + np.random.normal(0, 0.01, self.n_particles)
        self.particles[:, 2] = self.release_depth + np.random.normal(0, 5, self.n_particles)
        
        # All particles start at age 0, active, with concentration 1.0
        self.particles[:, 3] = 0  # age in days
        self.particles[:, 4] = 1  # active (1 = yes, 0 = no)
        self.particles[:, 5] = 1.0  # concentration
        
        print(f"âœ… Particles initialized at Day {self.simulation_day}")
        return self.particles
    
    def get_interpolated_current(self, lat, lon, depth):
        """Interpolate currents at particle position (simplified)"""
        if self.currents is None:
            # Fallback: simple gyre circulation
            return self.get_fallback_current(lat, lon, depth)
        
        # Simplified interpolation - in real code you'd use proper 3D interpolation
        u = np.sin(lon * np.pi / 180) * 0.1  # m/s
        v = np.cos(lat * np.pi / 180) * 0.1  # m/s
        w = 0.001 * np.sin(depth / 100)      # m/s
        
        return u, v, w
    
    def get_eke_at_position(self, lat, lon):
        """Get EKE value at position for diffusion enhancement"""
        if self.eke_data is None:
            return 0.01  # Default EKE
        
        # Simplified lookup - real code would interpolate
        return 0.05 + 0.1 * np.sin(lon * np.pi / 180) * np.cos(lat * np.pi / 180)
    
    def step(self):
        """Advance simulation by one time step"""
        if self.particles is None:
            self.initialize_particles()
        
        dt = self.time_step / self.substeps
        
        for substep in range(self.substeps):
            self._rk4_step(dt)
        
        # Update particle ages
        self.particles[:, 3] += self.time_step / (24 * 3600)  # Convert to days
        
        # Apply physics processes
        self._apply_diffusion()
        self._apply_vertical_mixing()
        self._apply_boundary_conditions()
        
        self.simulation_day += self.time_step / (24 * 3600)
    
    def _rk4_step(self, dt):
        """4th order Runge-Kutta integration (from your particleEngine.js)"""
        # Simplified RK4 implementation
        for i in range(self.n_particles):
            if self.particles[i, 4] == 0:  # Skip inactive particles
                continue
            
            lat, lon, depth = self.particles[i, :3]
            
            # k1
            u1, v1, w1 = self.get_interpolated_current(lat, lon, depth)
            
            # k2
            u2, v2, w2 = self.get_interpolated_current(
                lat + 0.5 * dt * v1 / 111000,  # Convert m/s to deg/s
                lon + 0.5 * dt * u1 / (111000 * np.cos(np.radians(lat))),
                depth + 0.5 * dt * w1
            )
            
            # k3
            u3, v3, w3 = self.get_interpolated_current(
                lat + 0.5 * dt * v2 / 111000,
                lon + 0.5 * dt * u2 / (111000 * np.cos(np.radians(lat))),
                depth + 0.5 * dt * w2
            )
            
            # k4 and final update
            u4, v4, w4 = self.get_interpolated_current(
                lat + dt * v3 / 111000,
                lon + dt * u3 / (111000 * np.cos(np.radians(lat))),
                depth + dt * w3
            )
            
            # Update position (weighted average of k1-k4)
            u = (u1 + 2*u2 + 2*u3 + u4) / 6
            v = (v1 + 2*v2 + 2*v3 + v4) / 6
            w = (w1 + 2*w2 + 2*w3 + w4) / 6
            
            # Apply advection
            self.particles[i, 0] += dt * v / 111000
            self.particles[i, 1] += dt * u / (111000 * np.cos(np.radians(lat)))
            self.particles[i, 2] += dt * w
    
    def _apply_diffusion(self):
        """Apply horizontal and vertical diffusion with EKE enhancement"""
        diffusion = self.params['diffusion_coefficient']
        vertical_diff = self.params['vertical_diffusion']
        
        for i in range(self.n_particles):
            if self.particles[i, 4] == 0:
                continue
            
            lat, lon, depth = self.particles[i, :3]
            
            # Get EKE-enhanced diffusion
            eke = self.get_eke_at_position(lat, lon)
            enhanced_diff = diffusion * (1 + self.params['eke_enhancement'] * eke)
            
            # Apply random walk (Brownian motion)
            self.particles[i, 0] += np.random.normal(0, np.sqrt(2 * enhanced_diff * self.time_step) / 111000)
            self.particles[i, 1] += np.random.normal(0, np.sqrt(2 * enhanced_diff * self.time_step) / 
                                                    (111000 * np.cos(np.radians(lat))))
            self.particles[i, 2] += np.random.normal(0, np.sqrt(2 * vertical_diff * self.time_step))
    
    def _apply_vertical_mixing(self):
        """Apply vertical mixing processes"""
        mixing = self.params['vertical_mixing']
        
        for i in range(self.n_particles):
            if self.particles[i, 4] == 0:
                continue
            
            # Random vertical displacement
            self.particles[i, 2] += np.random.normal(0, mixing)
            
            # Subsurface retention (less mixing at depth)
            if self.particles[i, 2] < -100:  # Below 100m
                self.particles[i, 2] *= self.params['subsurface_decay']
    
    def _apply_boundary_conditions(self):
        """Apply boundary conditions (surface, bottom, domain edges)"""
        for i in range(self.n_particles):
            if self.particles[i, 4] == 0:
                continue
            
            # Surface boundary (evaporation/loss)
            if self.particles[i, 2] > 0:  # Above surface
                self.particles[i, 2] = -1  # Push back to surface
                self.particles[i, 5] *= (1 - self.params['surface_evaporation'])
            
            # Bottom boundary (bathymetry)
            if self.bathymetry is not None:
                # Simplified: don't go below -4000m
                if self.particles[i, 2] < -4000:
                    self.particles[i, 2] = -3999
            
            # Domain boundaries (wrap longitude, limit latitude)
            if self.particles[i, 1] > 180:
                self.particles[i, 1] -= 360
            elif self.particles[i, 1] < -180:
                self.particles[i, 1] += 360
            
            if self.particles[i, 0] > 90:
                self.particles[i, 0] = 89.9
            elif self.particles[i, 0] < -90:
                self.particles[i, 0] = -89.9
            
            # Deactivate particles with very low concentration
            if self.particles[i, 5] < 0.01:
                self.particles[i, 4] = 0
    
    def get_statistics(self):
        """Get comprehensive particle statistics"""
        if self.particles is None:
            return {}
        
        active_mask = self.particles[:, 4] == 1
        active_particles = self.particles[active_mask]
        
        if len(active_particles) == 0:
            return {
                'n_active': 0,
                'mean_lat': 0,
                'mean_lon': 0,
                'mean_depth': 0,
                'spread_radius': 0,
                'max_depth': 0
            }
        
        stats = {
            'n_active': np.sum(active_mask),
            'mean_lat': np.mean(active_particles[:, 0]),
            'mean_lon': np.mean(active_particles[:, 1]),
            'mean_depth': np.mean(active_particles[:, 2]),
            'std_lat': np.std(active_particles[:, 0]),
            'std_lon': np.std(active_particles[:, 1]),
            'std_depth': np.std(active_particles[:, 2]),
            'spread_radius': np.sqrt(np.var(active_particles[:, 0]) + np.var(active_particles[:, 1])),
            'max_depth': np.min(active_particles[:, 2]),
            'min_concentration': np.min(active_particles[:, 5]),
            'max_concentration': np.max(active_particles[:, 5]),
            'simulation_day': self.simulation_day
        }
        
        return stats

# ============================================
# 2. PROFESSIONAL VISUALIZATION CLASS
# ============================================
print("\nðŸŽ¨ Creating ProfessionalVisualizer class...")

class ProfessionalVisualizer:
    """Complete visualization system for particle tracking research"""
    
    def __init__(self, engine):
        self.engine = engine
        self.fig = None
        self.axes = {}
        self.status_output = widgets.Output()
        
        # Create control panel
        self._create_control_panel()
        
        print("âœ… ProfessionalVisualizer initialized")
    
    def _create_control_panel(self):
        """Create interactive control panel"""
        # Parameter controls
        self.controls = {
            'diffusion_coefficient': widgets.FloatSlider(
                value=self.engine.params['diffusion_coefficient'],
                min=0.1,
                max=50.0,
                step=0.1,
                description='Diffusion (mÂ²/s):',
                continuous_update=False
            ),
            'eke_enhancement': widgets.FloatSlider(
                value=self.engine.params['eke_enhancement'],
                min=0.0,
                max=3.0,
                step=0.1,
                description='EKE Enhancement:',
                continuous_update=False
            ),
            'vertical_mixing': widgets.FloatSlider(
                value=self.engine.params['vertical_mixing'],
                min=0.0,
                max=1.0,
                step=0.01,
                description='Vertical Mixing:',
                continuous_update=False
            ),
            'simulation_speed': widgets.FloatSlider(
                value=1.0,
                min=0.1,
                max=10.0,
                step=0.1,
                description='Sim Speed:',
                continuous_update=False
            )
        }
        
        # Action buttons
        self.buttons = {
            'run_1day': widgets.Button(description='Run 1 Day', button_style='success'),
            'run_7days': widgets.Button(description='Run 7 Days', button_style='info'),
            'run_30days': widgets.Button(description='Run 30 Days', button_style='warning'),
                    'reset': widgets.Button(description='Reset Simulation', button_style='danger'),
            'export': widgets.Button(description='Export Data', button_style='primary')
        }
        
        # Visualization mode selector
        self.mode_selector = widgets.RadioButtons(
            options=['3d', '2d_map', 'vertical_section', 'statistics', 'all'],
            value='all',
            description='View:',
            disabled=False
        )
        
        # Connect button events
        self.buttons['run_1day'].on_click(lambda b: self.run_simulation(1))
        self.buttons['run_7days'].on_click(lambda b: self.run_simulation(7))
        self.buttons['run_30days'].on_click(lambda b: self.run_simulation(30))
        self.buttons['reset'].on_click(lambda b: self.reset_simulation())
        self.buttons['export'].on_click(lambda b: self.export_data())
        
        # Connect parameter updates
        for param, slider in self.controls.items():
            slider.observe(lambda change, p=param: self.update_parameter(p, change.new), 'value')
        
        # Create layout
        control_panel = widgets.VBox([
            widgets.HTML("<h3>ðŸ”¬ RESEARCHER CONTROL PANEL</h3>"),
            self.mode_selector,
            widgets.HTML("<hr>"),
            *self.controls.values(),
            widgets.HTML("<hr>"),
            widgets.HBox(list(self.buttons.values())),
            widgets.HTML("<hr>"),
            widgets.HTML("<h4>ðŸ“Š Simulation Status</h4>"),
            self.status_output
        ])
        
        # Create visualization area
        self._create_visualization_figure()
        
        # Display everything
        display(widgets.HBox([control_panel, widgets.Output()]))
        self.update_visualization()
    
    def _create_visualization_figure(self):
        """Create 6-panel visualization figure with Cartopy"""
        self.fig = plt.figure(figsize=(16, 12))
        
        # 1. 3D View
        self.axes['3d'] = self.fig.add_subplot(231, projection='3d')
        
        # 2. 2D Map with Cartopy
        self.axes['map'] = self.fig.add_subplot(232, projection=ccrs.PlateCarree())
        
        # 3. Vertical Section
        self.axes['vertical'] = self.fig.add_subplot(233)
        
        # 4. Statistics
        self.axes['stats'] = self.fig.add_subplot(234)
        
        # 5. EKE Field
        self.axes['eke'] = self.fig.add_subplot(235, projection=ccrs.PlateCarree())
        
        # 6. Histograms
        self.axes['hist'] = self.fig.add_subplot(236)
        
        plt.tight_layout()
    
    def update_parameter(self, param, value):
        """Update engine parameter"""
        if param in self.engine.params:
            self.engine.params[param] = value
            self.status_output.clear_output()
            with self.status_output:
                print(f"âœ… Updated {param}: {value}")
            self.update_visualization()
    
    def run_simulation(self, days):
        """Run simulation for specified days"""
        self.status_output.clear_output()
        with self.status_output:
            print(f"ðŸš€ Running {days} day simulation...")
        
        for day in range(days):
            self.engine.step()
            self.engine.simulation_day += 1
            
            # Update every 5 days for performance
            if day % 5 == 0:
                self.update_visualization()
                time.sleep(0.1)
        
        self.update_visualization()
        self.status_output.clear_output()
        with self.status_output:
            print(f"âœ… Completed {days} days. Day {self.engine.simulation_day}")
    
    def reset_simulation(self):
        """Reset simulation to initial state"""
        self.engine.initialize_particles()
        self.engine.simulation_day = 0
        self.update_visualization()
        self.status_output.clear_output()
        with self.status_output:
            print("ðŸ”„ Simulation reset to Day 0")
    
    def export_data(self):
        """Export particle data for analysis"""
        timestamp = datetime.now().strftime("%Y%m%d_%H%M%S")
        filename = f"particle_data_day{self.engine.simulation_day}_{timestamp}.nc"
        
        # Create xarray dataset
        ds = xr.Dataset({
            'lon': (['particle'], self.engine.particles[:, 0]),
            'lat': (['particle'], self.engine.particles[:, 1]),
            'depth': (['particle'], self.engine.particles[:, 2]),
            'age': (['particle'], self.engine.particles[:, 3])
        })
        
        # Add metadata
        ds.attrs['simulation_day'] = self.engine.simulation_day
        ds.attrs['n_particles'] = self.engine.n_particles
        ds.attrs['export_date'] = str(datetime.now())
        
        # Save to NetCDF
        ds.to_netcdf(filename)
        
        self.status_output.clear_output()
        with self.status_output:
            print(f"ðŸ’¾ Data exported to: {filename}")
            print(f"ðŸ“Š Particles: {self.engine.n_particles}")
            print(f"ðŸ“… Simulation day: {self.engine.simulation_day}")
    
    def update_visualization(self):
        """Update all visualization panels"""
        if self.fig is None:
            return
        
        # Clear all axes
        for ax in self.axes.values():
            ax.clear()
        
        # Get current mode
        mode = self.mode_selector.value
        
        # Update based on mode
        if mode == 'all' or mode == '3d':
            self._update_3d_view()
        
        if mode == 'all' or mode == '2d_map':
            self._update_2d_map()
        
        if mode == 'all' or mode == 'vertical_section':
            self._update_vertical_section()
        
        if mode == 'all' or mode == 'statistics':
            self._update_statistics()
        
        # Always show EKE and histograms
        self._update_eke_field()
        self._update_histograms()
        
        # Update title
        self.fig.suptitle(f'Fukushima Plume Simulation - Day {self.engine.simulation_day}', 
                         fontsize=16, y=1.02)
        
        plt.draw()
        self.fig.canvas.flush_events()
    
    def _update_3d_view(self):
        """Update 3D visualization"""
        ax = self.axes['3d']
        
        # Extract particle positions
        lons = self.engine.particles[:, 0]
        lats = self.engine.particles[:, 1]
        depths = self.engine.particles[:, 2]
        ages = self.engine.particles[:, 3]
        
        # Color by age
        colors = cm.viridis(ages / np.max(ages))
        
        # Scatter plot
        scatter = ax.scatter(lons, lats, depths, c=colors, s=1, alpha=0.6)
        
        # Set labels and limits
        ax.set_xlabel('Longitude (Â°E)')
        ax.set_ylabel('Latitude (Â°N)')
        ax.set_zlabel('Depth (m)')
        ax.set_title('3D Particle Distribution')
        
        # Set view angle
        ax.view_init(elev=30, azim=45)
        
        # Add colorbar for age
        sm = plt.cm.ScalarMappable(cmap=cm.viridis, 
                                  norm=plt.Normalize(vmin=0, vmax=np.max(ages)))
        sm.set_array([])
        plt.colorbar(sm, ax=ax, label='Particle Age (days)')
    
    def _update_2d_map(self):
        """Update 2D map with Cartopy"""
        ax = self.axes['map']
        
        # Add map features
        ax.add_feature(cfeature.LAND, facecolor='darkgreen', alpha=0.7)
        ax.add_feature(cfeature.OCEAN, facecolor='lightblue', alpha=0.3)
        ax.add_feature(cfeature.COASTLINE, linewidth=0.5)
        ax.add_feature(cfeature.BORDERS, linewidth=0.3, alpha=0.5)
        
        # Add gridlines
        ax.gridlines(draw_labels=True, linewidth=0.5, alpha=0.5)
        
        # Plot particles
        lons = self.engine.particles[:, 0]
        lats = self.engine.particles[:, 1]
        ages = self.engine.particles[:, 3]
        
        # Color by age
        colors = cm.viridis(ages / np.max(ages))
        
        scatter = ax.scatter(lons, lats, c=colors, s=2, alpha=0.6, 
                           transform=ccrs.PlateCarree())
        
        # Mark Fukushima location
        ax.plot(141.03, 37.42, 'r*', markersize=10, 
               transform=ccrs.PlateCarree(), label='Fukushima')
        
        # Set extent
        ax.set_extent([120, 180, 20, 50], crs=ccrs.PlateCarree())
        ax.set_title('Surface Distribution')
        ax.legend()
    
    def _update_vertical_section(self):
        """Update vertical section view"""
        ax = self.axes['vertical']
        
        # Extract data
        lons = self.engine.particles[:, 0]
        depths = self.engine.particles[:, 2]
        ages = self.engine.particles[:, 3]
        
        # Create vertical section
        scatter = ax.scatter(lons, depths, c=ages, cmap='viridis', s=1, alpha=0.6)
        
        # Add bathymetry if available
        if self.engine.bathymetry is not None:
            # Sample bathymetry along particle longitudes
            pass
        
        ax.set_xlabel('Longitude (Â°E)')
        ax.set_ylabel('Depth (m)')
        ax.set_title('Vertical Section')
        ax.invert_yaxis()  # Depth increases downward
        ax.grid(True, alpha=0.3)
        
        # Add colorbar
        plt.colorbar(scatter, ax=ax, label='Age (days)')
    
    def _update_statistics(self):
        """Update statistics panel"""
        ax = self.axes['stats']
        
        # Calculate statistics
        lons = self.engine.particles[:, 0]
        lats = self.engine.particles[:, 1]
        depths = self.engine.particles[:, 2]
        ages = self.engine.particles[:, 3]
        
        stats_text = f"""
        SIMULATION STATISTICS
        =====================
        Day: {self.engine.simulation_day}
        Particles: {self.engine.n_particles}
        
        Position Statistics:
        --------------------
        Longitude: {lons.mean():.2f}Â°E Â± {lons.std():.2f}
        Latitude: {lats.mean():.2f}Â°N Â± {lats.std():.2f}
        Depth: {depths.mean():.1f}m Â± {depths.std():.1f}
        
        Age Statistics:
        ---------------
        Mean Age: {ages.mean():.1f} days
        Max Age: {ages.max():.1f} days
        Min Age: {ages.min():.1f} days
        
        Spread Metrics:
        ---------------
        Zonal Spread: {lons.max() - lons.min():.1f}Â°
        Meridional Spread: {lats.max() - lats.min():.1f}Â°
        Vertical Spread: {depths.max() - depths.min():.1f}m
        """
        
        ax.text(0.05, 0.95, stats_text, transform=ax.transAxes,
               fontfamily='monospace', fontsize=9, verticalalignment='top',
               bbox=dict(boxstyle='round', facecolor='lightblue', alpha=0.3))
        
        ax.set_title('Statistics Dashboard')
        ax.axis('off')
    
    def _update_eke_field(self):
        """Update EKE field visualization"""
        ax = self.axes['eke']
        
        if self.engine.eke_data is None:
            ax.text(0.5, 0.5, 'EKE Data Not Loaded', 
                   transform=ax.transAxes, ha='center', va='center')
            ax.set_title('EKE Field (Not Available)')
            return
        
        # Add map features
        ax.add_feature(cfeature.LAND, facecolor='darkgreen', alpha=0.7)
        ax.add_feature(cfeature.COASTLINE, linewidth=0.5)
        
        # Plot EKE field (simplified - you'd use actual EKE data here)
        # This is a placeholder - you'd load your actual CMEMS EKE data
        lons_grid = np.linspace(120, 180, 50)
        lats_grid = np.linspace(20, 50, 50)
        lon_grid, lat_grid = np.meshgrid(lons_grid, lats_grid)
        
        # Create synthetic EKE field (replace with your actual data)
        eke_field = np.exp(-((lon_grid - 150)**2 + (lat_grid - 35)**2) / 500)
        
        # Contour plot
        contour = ax.contourf(lon_grid, lat_grid, eke_field, 
                             transform=ccrs.PlateCarree(),
                             cmap='YlOrRd', alpha=0.7)
        
        # Overlay particles
        lons = self.engine.particles[:, 0]
        lats = self.engine.particles[:, 1]
        ax.scatter(lons, lats, c='blue', s=1, alpha=0.3,
                  transform=ccrs.PlateCarree())
        
        ax.set_extent([120, 180, 20, 50], crs=ccrs.PlateCarree())
        ax.set_title('EKE Field with Particles')
        
        # Add colorbar
        plt.colorbar(contour, ax=ax, label='EKE (mÂ²/sÂ²)')
    
    def _update_histograms(self):
        """Update histogram panel"""
        ax = self.axes['hist']
        
        # Create subplots within the histogram panel
        ax.clear()
        
        # Create 2x2 grid of histograms
        gs = ax._subplotspec.subgridspec(2, 2)
        
        # 1. Age distribution
        ax1 = self.fig.add_subplot(gs[0, 0])
        ages = self.engine.particles[:, 3]
        ax1.hist(ages, bins=30, color='skyblue', edgecolor='black', alpha=0.7)
        ax1.set_xlabel('Age (days)')
        ax1.set_ylabel('Count')
        ax1.set_title('Age Distribution')
        ax1.grid(True, alpha=0.3)
        
        # 2. Depth distribution
        ax2 = self.fig.add_subplot(gs[0, 1])
        depths = self.engine.particles[:, 2]
        ax2.hist(depths, bins=30, color='lightgreen', edgecolor='black', alpha=0.7)
        ax2.set_xlabel('Depth (m)')
        ax2.set_ylabel('Count')
        ax2.set_title('Depth Distribution')
        ax2.invert_xaxis()  # Depth increases to left
        ax2.grid(True, alpha=0.3)
        
        # 3. Latitude distribution
        ax3 = self.fig.add_subplot(gs[1, 0])
        lats = self.engine.particles[:, 1]
        ax3.hist(lats, bins=30, color='salmon', edgecolor='black', alpha=0.7)
        ax3.set_xlabel('Latitude (Â°N)')
        ax3.set_ylabel('Count')
        ax3.set_title('Latitude Distribution')
        ax3.grid(True, alpha=0.3)
        
        # 4. Longitude distribution
        ax4 = self.fig.add_subplot(gs[1, 1])
        lons = self.engine.particles[:, 0]
        ax4.hist(lons, bins=30, color='gold', edgecolor='black', alpha=0.7)
        ax4.set_xlabel('Longitude (Â°E)')
        ax4.set_ylabel('Count')
        ax4.set_title('Longitude Distribution')
        ax4.grid(True, alpha=0.3)
        
        # Remove the original axis
        ax.remove()

# ============================================
# 3. QUICK START FUNCTION
# ============================================
def quick_start():
    """Quick start for immediate visualization"""
    print("\nðŸš€ Launching Quick Start...")
    
    # Create engine with default parameters
    engine = ParticleEngine3D(n_particles=5000)
    engine.initialize_particles()
    
    # Create visualizer
    visualizer = ProfessionalVisualizer(engine)
    
    print("âœ… Quick Start ready! Use the control panel to run simulations.")
    return visualizer

# ============================================
# 4. RESEARCH MODE WITH YOUR DATA
# ============================================
def launch_research_visualizer():
    """Full research mode with your actual data"""
    print("\nðŸ”¬ Launching Research Visualizer...")
    
    # Create engine
    engine = ParticleEngine3D(n_particles=10000)
    
    # Load your actual data
    print("ðŸ“‚ Loading your data files...")
    try:
        # Load HYCOM currents
        print("  - Loading HYCOM currents...")
        # engine.load_currents('data/hycom_currents_2011_2013.nc')
        
        # Load CMEMS EKE
        print("  - Loading CMEMS EKE data...")
        # engine.load_eke('data/cmems_eke_2011_2013.nc')
        
        # Load bathymetry
        print("  - Loading bathymetry...")
        # engine.load_bathymetry('data/gebco_bathymetry.nc')
        
        print("âœ… Data loaded successfully!")
    except Exception as e:
        print(f"âš ï¸  Could not load data files: {e}")
        print("   Using synthetic data for demonstration.")
    
    # Initialize particles
    engine.initialize_particles()
    
    # Create visualizer
    visualizer = ProfessionalVisualizer(engine)
    
    print("âœ… Research Visualizer ready!")
    print("   Use the control panel to adjust parameters and run simulations.")
    
    return visualizer

# ============================================
# 5. MAIN EXECUTION
# ============================================
if __name__ == "__main__":
    print("\n" + "="*80)
    print("ðŸŒŠ FUKUSHIMA PLUME 3D VISUALIZATION SYSTEM")
    print("MODERN VERSION WITH CARTOPY")
    print("="*80)
    
    print("\nChoose mode:")
    print("1. Quick Start (immediate visualization)")
    print("2. Research Mode (with your data files)")
    print("3. Create Research Report")
    
    choice = input("\nEnter choice (1-3): ").strip()
    
    if choice == "1":
        visualizer = quick_start()
    elif choice == "2":
        visualizer = launch_research_visualizer()
    elif choice == "3":
        # Create research report
        engine = ParticleEngine3D(n_particles=5000)
        engine.initialize_particles()
        # create_research_report(engine)  # You can implement this
        print("ðŸ“„ Research report function to be implemented")
    else:
        print("âŒ Invalid choice. Running quick start...")
        visualizer = quick_start()

print("\n" + "="*80)
print("âœ… PROFESSIONAL 3D VISUALIZATION SYSTEM READY")
print("="*80)
print("\nTo use in your Jupyter notebook:")
print("1. Copy this entire script into a cell")
print("2. Run: visualizer = quick_start()  # For immediate use")
print("3. Or: visualizer = launch_research_visualizer()  # With your data")
print("\nHappy researching! ðŸš€")
```

```python
import os
import glob
import numpy as np

def diagnose_bin_files():
    """Check what's in your .bin files"""
    print("ðŸ” DIAGNOSING .BIN FILES...")
    
    # Change this to your .bin directory
    bin_dir = "/Users/shuian/PycharmProjects/Fukushima_Plume_Simulator/data/eke_ultra_optimized/daily/"
    
    # Find all .bin files
    bin_files = sorted(glob.glob(os.path.join(bin_dir, "*.bin")))
    
    if not bin_files:
        print(f"âŒ No .bin files found in {bin_dir}")
        # Try other patterns
        print("   Searching for other patterns...")
        bin_files = sorted(glob.glob(os.path.join(bin_dir, "*")))
        print(f"   All files: {bin_files[:10]}...")
        return
    
    print(f"âœ… Found {len(bin_files)} .bin files")
    print(f"   First 5: {[os.path.basename(f) for f in bin_files[:5]]}")
    
    # Check first file
    first_file = bin_files[0]
    print(f"\nðŸ“Š Analyzing: {os.path.basename(first_file)}")
    
    # Get file size
    file_size = os.path.getsize(first_file)
    print(f"   File size: {file_size} bytes")
    
    # Try to read with different data types
    for dtype in [np.float32, np.float64, np.int32]:
        try:
            data = np.fromfile(first_file, dtype=dtype)
            print(f"   As {dtype}: {len(data)} elements")
            
            if len(data) > 0:
                print(f"      Shape guess: Try reshaping to...")
                
                # Common grid sizes
                possible_shapes = [
                    (450, 1125),    # HYCOM 0.08Â°
                    (400, 1040),    # Original 0.125Â°
                    (720, 1440),    # 0.25Â°
                    (361, 720),     # 0.5Â°
                ]
                
                for shape in possible_shapes:
                    if shape[0] * shape[1] == len(data):
                        print(f"      âœ… {shape[0]}Ã—{shape[1]} = {len(data)}")
                        # Reshape and show stats
                        data_2d = data.reshape(shape)
                        print(f"         Min: {data_2d.min():.2e}, Max: {data_2d.max():.2e}")
                        print(f"         Mean: {data_2d.mean():.2e}, Std: {data_2d.std():.2e}")
                        break
                
                break  # Stop if we found a valid dtype
                
        except Exception as e:
            continue
    
    return bin_files

# Run diagnosis
bin_files = diagnose_bin_files()
```

```python
import xarray as xr
import numpy as np
from scipy.interpolate import griddata

def interpolate_eke_to_3d_hycom():
    """Interpolate CMEMS EKE to HYCOM's 3D curvilinear grid"""
    print("ðŸ“ Interpolating EKE to 3D HYCOM grid...")
    
    # Paths
    CMEMS_PATH = "/Users/shuian/PycharmProjects/Fukushima_Plume_Simulator/archive/cmems_EKE_data/cmems_obs-sl_glo_phy-ssh_my_allsat-l4-duacs-0.125deg_P1D_1769906652641.nc"
    HYCOM_PATH = "/Users/shuian/PycharmProjects/Fukushima_Plume_Simulator/data/hycom_combined_march/hycom_march_2011_combined.nc"
    OUTPUT_PATH = "/Users/shuian/PycharmProjects/Fukushima_Plume_Simulator/archive/cmems_EKE_data/eke_interpolated_to_3d_hycom.nc"
    
    # Load data
    print(f"ðŸ“¥ Loading CMEMS...")
    cmems = xr.open_dataset(CMEMS_PATH)
    
    print(f"ðŸ“¥ Loading HYCOM...")
    hycom = xr.open_dataset(HYCOM_PATH)
    
    # Get HYCOM's 2D surface grid
    hycom_lat_2d = hycom.Latitude.values  # Shape: (1261, 1626)
    hycom_lon_2d = hycom.Longitude.values
    
    print(f"\nðŸ“Š Grid Information:")
    print(f"   CMEMS: {len(cmems.latitude)}Ã—{len(cmems.longitude)} (regular)")
    print(f"   HYCOM: {hycom_lat_2d.shape} (curvilinear)")
    print(f"   HYCOM depths: {hycom.Depth.values} m")
    print(f"   HYCOM time steps: {len(hycom.MT)}")
    
    # Calculate EKE from CMEMS
    print("\nðŸ§® Calculating EKE from ugosa/vgosa...")
    cmems_eke = 0.5 * (cmems['ugosa']**2 + cmems['vgosa']**2)
    
    # Prepare grids for interpolation
    cmems_lon, cmems_lat = np.meshgrid(cmems.longitude.values, cmems.latitude.values)
    cmems_points = np.column_stack([cmems_lon.ravel(), cmems_lat.ravel()])
    hycom_points = np.column_stack([hycom_lon_2d.ravel(), hycom_lat_2d.ravel()])
    
    # Match time steps
    cmems_times = cmems.time.values
    hycom_times = hycom.MT.values
    
    print(f"\nâ° Time alignment:")
    print(f"   CMEMS: {len(cmems_times)} steps ({cmems_times[0]} to {cmems_times[-1]})")
    print(f"   HYCOM: {len(hycom_times)} steps")
    
    # Interpolate each time step
    eke_interpolated = []
    
    for i, hycom_time in enumerate(hycom_times):
        print(f"   Time {i+1}/{len(hycom_times)}...", end='\r')
        
        # Find closest CMEMS time
        time_idx = np.argmin(np.abs(cmems_times - hycom_time))
        
        # Get EKE for this time
        eke_flat = cmems_eke[time_idx].values.ravel()
        
        # Interpolate to HYCOM grid
        eke_interp = griddata(
            cmems_points, eke_flat, hycom_points,
            method='linear', fill_value=0.0
        ).reshape(hycom_lat_2d.shape)
        
        eke_interpolated.append(eke_interp)
    
    print(f"\nâœ… Interpolated {len(hycom_times)} time steps")
    
    # Create 3D EKE array (same for all depths)
    eke_3d = np.array(eke_interpolated)  # Shape: (time, Y, X)
    
    # Expand to 3D (same EKE for all depths)
    eke_3d_expanded = np.repeat(eke_3d[:, np.newaxis, :, :], len(hycom.Depth), axis=1)
    
    print(f"   Final shape: {eke_3d_expanded.shape} (MT, Depth, Y, X)")
    
    # Create dataset
    ds = xr.Dataset({
        'eke': (['MT', 'Depth', 'Y', 'X'], eke_3d_expanded)
    }, coords={
        'MT': hycom.MT.values,
        'Depth': hycom.Depth.values,
        'Y': hycom.Y.values,
        'X': hycom.X.values,
        'Latitude': (['Y', 'X'], hycom_lat_2d),
        'Longitude': (['Y', 'X'], hycom_lon_2d)
    })
    
    ds.eke.attrs = {
        'units': 'mÂ²/sÂ²',
        'long_name': 'Eddy Kinetic Energy',
        'source': 'CMEMS ugosa/vgosa interpolated to HYCOM grid'
    }
    
    ds.to_netcdf(OUTPUT_PATH)
    print(f"âœ… Saved to: {OUTPUT_PATH}")
    
    # Quick verification
    print(f"\nðŸ“Š Verification:")
    print(f"   EKE min: {ds.eke.min().values:.2e} mÂ²/sÂ²")
    print(f"   EKE max: {ds.eke.max().values:.2e} mÂ²/sÂ²")
    print(f"   EKE mean: {ds.eke.mean().values:.2e} mÂ²/sÂ²")
    
    return ds

# Run interpolation
eke_3d_data = interpolate_eke_to_3d_hycom()
```

```python
import xarray as xr

def debug_hycom_structure():
    """Check what's actually in your HYCOM file"""
    HYCOM_PATH = "/Users/shuian/PycharmProjects/Fukushima_Plume_Simulator/data/hycom_combined_march/hycom_march_2011_combined.nc"
    
    print(f"ðŸ” Opening: {HYCOM_PATH}")
    ds = xr.open_dataset(HYCOM_PATH)
    
    print("\nðŸ“Š FILE STRUCTURE:")
    print("="*60)
    
    # Dimensions
    print("DIMENSIONS:")
    for dim, size in ds.dims.items():
        print(f"  {dim}: {size}")
    
    # Variables
    print("\nVARIABLES:")
    for var in ds.variables:
        print(f"  {var}: {ds[var].dims} {ds[var].shape}")
        if hasattr(ds[var], 'units'):
            print(f"     units: {ds[var].units}")
    
    # Coordinates (likely candidates for lat/lon)
    print("\nCOORDINATE CANDIDATES:")
    coord_candidates = ['lat', 'latitude', 'Lat', 'Latitude', 
                       'lon', 'longitude', 'Lon', 'Longitude',
                       'x', 'y', 'i', 'j', 'nav_lat', 'nav_lon']
    
    for candidate in coord_candidates:
        if candidate in ds.variables:
            var = ds[candidate]
            print(f"  âœ… {candidate}: {var.shape} ({var.attrs.get('units', 'no units')})")
            if len(var.shape) == 1:
                print(f"     range: {var.values[0]:.2f} to {var.values[-1]:.2f}")
            elif len(var.shape) == 2:
                print(f"     min: {var.values.min():.2f}, max: {var.values.max():.2f}")
    
    # Time information
    print("\nTIME INFORMATION:")
    time_vars = ['time', 'Time', 't', 'T']
    for tvar in time_vars:
        if tvar in ds.variables:
            print(f"  âœ… {tvar}: {len(ds[tvar])} time steps")
            print(f"     first: {ds[tvar].values[0]}")
            print(f"     last: {ds[tvar].values[-1]}")
    
    return ds

# Run debug
hycom_ds = debug_hycom_structure()
```
