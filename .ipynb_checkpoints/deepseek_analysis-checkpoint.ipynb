```python
# Cell: Complete solution using absolute paths

import os
from pathlib import Path
from dotenv import load_dotenv
import requests
from IPython.display import Markdown, display

# === SETUP PATHS ===
NOTEBOOK_DIR = Path(".").absolute()
PROJECT_ROOT = NOTEBOOK_DIR.parent

print(f"ðŸ““ Notebook: {NOTEBOOK_DIR}")
print(f"ðŸ“ Project: {PROJECT_ROOT}")

# === LOAD API KEY ===
# Load .env from project root
env_path = PROJECT_ROOT / ".env"
if env_path.exists():
    load_dotenv(env_path)
    print(f"âœ… Loaded .env from: {env_path}")
else:
    load_dotenv()
    print("âš  Using default .env location")

API_KEY = os.getenv("DEEPSEEK_API_KEY")

if not API_KEY:
    print("âŒ Set DEEPSEEK_API_KEY in .env file!")
    exit()

print(f"âœ… API key loaded")

# === FILE LIST ===
FILE_PATHS = [
    PROJECT_ROOT / "web/app.js",
    PROJECT_ROOT / "web/index.html",
    PROJECT_ROOT / "web/particleEngine.js",
    PROJECT_ROOT / "web/hycomLoader.js",
    PROJECT_ROOT / "web/cmemsLoader.js",
    PROJECT_ROOT / "web/coordinateSystem3D.js",
    PROJECT_ROOT / "web/rk4Solver3D.js",
    PROJECT_ROOT / "web/hycomWeb3DLoader.js",
    PROJECT_ROOT / "web/verticalInterpolator.js",
    PROJECT_ROOT / "scripts/CMEMS_Heatmap_Analysis.py",
    PROJECT_ROOT / "scripts/HYCOM_Heatmap_Analysis.py",
    PROJECT_ROOT / "scripts/download_HYCOM.py",
    PROJECT_ROOT / "scripts/prepare_CMEMS.py",
    PROJECT_ROOT / "scripts/prepare_HYCOM.py",
    PROJECT_ROOT / "scripts/prepare_HYCOM3D.py",
    PROJECT_ROOT / "README.md",
    PROJECT_ROOT / "requirements.txt"
]

print(f"\nðŸ” Checking files...")
existing_files = []
for path in FILE_PATHS:
    if path.exists():
        size = path.stat().st_size
        existing_files.append((path, size))
        print(f"  âœ“ {path.relative_to(PROJECT_ROOT)} ({size:,} bytes)")
    else:
        print(f"  âš  Missing: {path.relative_to(PROJECT_ROOT)}")

print(f"\nâœ… Found {len(existing_files)}/{len(FILE_PATHS)} files")

# === LOAD CONTEXT ===
_CONTEXT_CACHE = None

def load_context():
    """Load context using absolute paths"""
    global _CONTEXT_CACHE
    
    if _CONTEXT_CACHE is not None:
        print("ðŸ“‚ Using cached context")
        return _CONTEXT_CACHE
    
    print("ðŸ“‚ Loading context...")
    
    context_parts = []
    total_chars = 0
    
    for path, size in existing_files:
        try:
            with open(path, 'r', encoding='utf-8') as f:
                content = f.read()
            
            # Trim if large
            if len(content) > 8000:
                content = content[:6000] + f"\n... [truncated, original {len(content):,} chars] ..."
            
            rel_path = path.relative_to(PROJECT_ROOT)
            context_parts.append(f"=== {rel_path} ===\n{content}\n")
            total_chars += len(content)
            
        except Exception as e:
            print(f"  âŒ Error reading {path.name}: {str(e)[:50]}")
    
    _CONTEXT_CACHE = "\n".join(context_parts)
    
    print(f"\nâœ… Context loaded:")
    print(f"   Files: {len(context_parts)}")
    print(f"   Size: {total_chars:,} chars ({(total_chars/1024):.1f} KB)")
    
    return _CONTEXT_CACHE

# Load it
context = load_context()
print("\nðŸŽ‰ Ready to ask questions!")
```

```python
# Cell 2: Core question-asking function

import requests
import time
from IPython.display import Markdown, display

def ask(question, temperature=0.7, max_tokens=3000):
    """
    Ask DeepSeek a question using your cached context
    """
    global _CONTEXT_CACHE
    
    # Ensure context is loaded
    if _CONTEXT_CACHE is None:
        context = load_context()
    else:
        context = _CONTEXT_CACHE
    
    print(f"ðŸ¤” Question: {question}")
    print(f"ðŸ“Š Using {len(context):,} chars of context")
    
    # Prepare the API request
    headers = {"Authorization": f"Bearer {API_KEY}"}
    
    messages = [
        {
            "role": "system",
            "content": f"""You are an expert in ocean modeling and particle tracking.
You're assisting with a Fukushima radioactive plume simulation.

Here's the complete codebase:
{context}

IMPORTANT: Provide SPECIFIC, ACTIONABLE advice. Reference actual file names and functions.
When suggesting code changes, show EXACT code snippets that would work with the existing codebase."""
        },
        {"role": "user", "content": question}
    ]
    
    print(f"â³ Sending to DeepSeek...")
    start_time = time.time()
    
    try:
        response = requests.post(
            "https://api.deepseek.com/v1/chat/completions",
            headers=headers,
            json={
                "model": "deepseek-chat",
                "messages": messages,
                "temperature": temperature,
                "max_tokens": max_tokens,
                "stream": False
            },
            timeout=90  # 90 seconds for complex questions
        )
        
        elapsed = time.time() - start_time
        
        if response.status_code == 200:
            answer = response.json()['choices'][0]['message']['content']
            print(f"âœ… Answer received in {elapsed:.1f}s")
            print("="*70)
            display(Markdown(answer))
            print("="*70)
            
            # Save to history
            save_to_history(question, answer)
            
            return answer
        else:
            print(f"âŒ API Error {response.status_code}: {response.text[:200]}")
            return None
            
    except requests.exceptions.Timeout:
        print(f"â± Timeout after 90s")
        return None
    except Exception as e:
        print(f"âŒ Error: {type(e).__name__}: {str(e)[:100]}")
        return None

def save_to_history(question, answer):
    """Save Q&A to markdown file"""
    import datetime
    
    history_file = NOTEBOOK_DIR / "assistant_history.md"
    
    with open(history_file, "a", encoding="utf-8") as f:
        timestamp = datetime.datetime.now().strftime("%Y-%m-%d %H:%M:%S")
        f.write(f"\n## {timestamp}\n")
        f.write(f"**Q:** {question}\n\n")
        f.write(f"**A:** {answer}\n\n")
        f.write("---\n")
    
    print(f"ðŸ’¾ Saved to: {history_file}")

print("âœ… Question function ready!")
print("ðŸ’¡ Use: ask('Your question here')")
```

```python
# Cell: Your Complete "One-Click" DeepSeek Assistant
import os
import json
import time
from pathlib import Path
from dotenv import load_dotenv
import requests
from IPython.display import Markdown, display

# === 1. LOAD CONFIGURATION ===
load_dotenv()
API_KEY = os.getenv("DEEPSEEK_API_KEY")
MODEL = "deepseek-chat"  # Default model; switch to "deepseek-reasoner" for complex reasoning[citation:1]
API_BASE = "https://api.deepseek.com/v1"

if not API_KEY:
    print("âŒ Please set your DEEPSEEK_API_KEY in a .env file or directly in this cell.")
    # You can set it temporarily for testing:
    # API_KEY = "your_key_here"

# === 2. DEFINE YOUR PROJECT FILES ===
PROJECT_ROOT = Path.cwd()  # Assumes your notebook is in the project root
FILE_PATHS = [
    PROJECT_ROOT / "web/app.js",
    PROJECT_ROOT / "web/particleEngine.js",
    PROJECT_ROOT / "web/rk4Solver3D.js",
    PROJECT_ROOT / "web/hycomLoader.js",
    PROJECT_ROOT / "scripts/prepare_HYCOM3D.py",
    PROJECT_ROOT / "README.md",
    # Add all other relevant files from your project here
]

# === 3. SET UP THE EFFICIENT ASSISTANT ===

# Global caches to load files and conversation only once
_CONTEXT_CACHE = None
_CONVERSATION_HISTORY = []
_MAX_HISTORY = 6  # Keeps the last 6 exchanges for context

def load_project_context_once():
    """Loads all project files into memory, runs only once per session."""
    global _CONTEXT_CACHE
    if _CONTEXT_CACHE is not None:
        return _CONTEXT_CACHE  # Use cache if already loaded

    print("ðŸ“‚ Loading project context (this happens once)...")
    context_parts = []
    total_size = 0

    for file_path in FILE_PATHS:
        try:
            if file_path.exists():
                with open(file_path, 'r', encoding='utf-8') as f:
                    content = f.read()
                # Smart truncation: Keep file structure but limit very long files
                if len(content) > 8000:
                    content = f"{content[:6000]}\n\n... [file truncated for brevity] ...\n\n{content[-1500:]}"
                rel_path = file_path.relative_to(PROJECT_ROOT)
                context_parts.append(f"--- File: {rel_path} ---\n{content}")
                total_size += len(content)
                print(f"  âœ“ {rel_path}")
            else:
                print(f"  âš  Not found: {file_path.name}")
        except Exception as e:
            print(f"  âœ— Error reading {file_path.name}: {e}")

    _CONTEXT_CACHE = "\n\n".join(context_parts)
    print(f"âœ… Context ready: {len(_CONTEXT_CACHE):,} characters from {len(context_parts)} files.\n")
    return _CONTEXT_CACHE

def ask_assistant(question, use_history=True, temperature=0.7, thinking_mode=False):
    """
    Your main function. Call this with any question.
    It uses cached files, manages conversation memory, and displays the answer.

    Args:
        question (str): Your question.
        use_history (bool): If True, includes past Q&A for context. Set to False for isolated queries.
        temperature (float): Controls creativity (0.0 = precise, 1.0 = creative).
        thinking_mode (bool): If True, uses 'deepseek-reasoner' for step-by-step reasoning[citation:1][citation:7].
    """
    # A. Load Code Context
    code_context = load_project_context_once()

    # B. Manage Conversation History
    if use_history:
        # Build the message list with system prompt and history
        messages = [
            {"role": "system", "content": f"You are a coding assistant for an ocean particle simulation. Use this codebase as reference:\n\n{code_context[:12000]}"}
        ]
        # Add the last few turns of conversation
        for exchange in _CONVERSATION_HISTORY[-_MAX_HISTORY:]:
            messages.append({"role": "user", "content": exchange["question"]})
            messages.append({"role": "assistant", "content": exchange["answer"][:500]})  # Truncated history
        # Add the current question
        messages.append({"role": "user", "content": question})
    else:
        # Fresh conversation without history
        messages = [
            {"role": "system", "content": f"Codebase for context:\n\n{code_context[:12000]}"},
            {"role": "user", "content": question}
        ]

    # C. Call the API
    headers = {"Authorization": f"Bearer {API_KEY}", "Content-Type": "application/json"}
    chosen_model = "deepseek-reasoner" if thinking_mode else MODEL[citation:1]

    payload = {
        "model": chosen_model,
        "messages": messages,
        "temperature": temperature,
        "max_tokens": 4000,  # Increase for very long answers
        "stream": False  # Set to True for word-by-word streaming
    }

    print(f"ðŸ¤” Processing your question...")
    start_time = time.time()

    try:
        response = requests.post(f"{API_BASE}/chat/completions", headers=headers, json=payload, timeout=90)
        elapsed = time.time() - start_time

        if response.status_code == 200:
            answer = response.json()['choices'][0]['message']['content']

            # D. Update History and Display
            if use_history:
                _CONVERSATION_HISTORY.append({"question": question, "answer": answer})
                if len(_CONVERSATION_HISTORY) > _MAX_HISTORY:
                    _CONVERSATION_HISTORY.pop(0)  # Keep list size managed

            print(f"âœ… Answer ready ({elapsed:.1f}s)")
            display(Markdown(f"**Answer:**\n\n{answer}"))
            return answer
        else:
            print(f"âŒ API Error {response.status_code}: {response.text[:200]}")
            return None

    except requests.exceptions.Timeout:
        print("â± Request timed out. The question or context might be very complex.")
        return None
    except Exception as e:
        print(f"âŒ Request failed: {type(e).__name__}")
        return None

# === 4. HELPER FUNCTIONS FOR YOUR WORKFLOW ===
def clear_history():
    """Clear the conversation memory."""
    global _CONVERSATION_HISTORY
    _CONVERSATION_HISTORY = []
    print("ðŸ—‘ï¸ Conversation history cleared.")

def show_history():
    """Show the recent conversation."""
    if not _CONVERSATION_HISTORY:
        print("No conversation history yet.")
        return
    print("ðŸ“œ Recent Conversation:")
    for i, exchange in enumerate(_CONVERSATION_HISTORY, 1):
        print(f"  {i}. Q: {exchange['question'][:60]}...")
        print(f"     A: {exchange['answer'][:60]}...\n")

# === READY TO USE ===
print("ðŸš€ Your efficient DeepSeek Assistant is ready!")
print("ðŸ’¡ Main command: answer = ask_assistant('Your question here')")
print("âš™ï¸  Helpers: clear_history(), show_history()")
print("---")
# Quick test - uncomment the line below to try it immediately
# test_answer = ask_assistant("What is the main function of my particleEngine.js file?", use_history=False)
```
